```python
# get scoring_offense, scoring_defense, player_APG, and player RPG for everyteam
# team scoring offense
def get_team_defense():
    browser = Browser('chrome')
    time.sleep(1)
    url = 'https://www.ncaa.com/stats/basketball-men/d1'
    browser.visit(url)
    time.sleep(1)
    # from homepage, click on scoring offense link
    browser.links.find_by_partial_text('Scoring Defense').click()
    time.sleep(1)
    counter = 2

    html = browser.html
    scoring_defense_soup = bs(html)

    pages_list = scoring_defense_soup.find_all('li', class_ = 'stats-pager__li')
    pages = len(pages_list) - 1

    for x in range(pages):
        #scrape table from first page then go to next page
        html = browser.html
        table = pd.read_html(html)

        if x > 0:
            df2 = table[0]
            df = pd.concat([df,df2]).reset_index().drop(columns = 'index')
        else:
            df = table[0]

        try:
            browser.links.find_by_href(f"/stats/basketball-men/d1/current/team/146/p{counter}").click()
            time.sleep(1)
        except:
            print('Last page reached for team defense..')

        counter +=1
    browser.quit()
    return df
#scrape table from first page then go to next page
```

```python
def merge_dfs(df1,df2,df3,df4,df5,df6):
    df_list = [df3,df4,df5]
    
    merged_df = pd.merge(df1,df2, on = ['Team','GM']).drop(columns = ['Rank_x','Rank_y'])
    
    for df in df_list:
        merged_df = pd.merge(merged_df, df,on = ['Team', "GM"]).drop(columns = 'Rank')
        
    merged_df = pd.merge(merged_df,df6, on = ['Team']).drop(columns = 'Rank')
    merged_df = merged_df.rename(columns = {'PPG_x':'PPG','PPG_y':'Bench_PPG'})
    
    return merged_df
```

```python
def get_stats():
    team_scoring_offense = get_team_offense()
    team_scoring_defense = get_team_defense()
    team_assists = get_team_assists()
    team_rebounds = get_team_rebounds()
    team_bench_points = get_team_bench_points()
    team_win_percent = get_team_win_percent()
    
    merged_df = merge_dfs(
                        team_scoring_offense,
                        team_scoring_defense,
                        team_assists,
                        team_rebounds,
                        team_bench_points,
                        team_win_percent
                        )
    
    return merged_df
```
